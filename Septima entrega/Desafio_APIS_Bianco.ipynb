## Desarrollo

Primer se importan las librerias necesarias para trabajar.

```python
import requests
import json
import pandas as pd
```

Luego de haber importado estas librerias nos conectamos a la API y extraemos los datos que nos proporciona, para se debe generar una key la cual ya se ha obtenido previamente desde la web.

```python
url = "https://starcraft-ii.p.rapidapi.com/people/page/1/"
```

```python
headers = {
    "X-RapidAPI-Key": "8043e14394msh151162b4501a4bfp10b13ajsn63e36d6c38fa",
    "X-RapidAPI-Host": "starcraft-ii.p.rapidapi.com"
}
```

```python
response = requests.request("GET", url, headers=headers)
response
```

Vemos que el codigo obtenido es 200, lo que significa que la solicitud se realizo con exito y que tenemos los datos en la variable response.

A continuacion transformaremos estos datos a un dataframe utilizando la libreria json

```python
jsondata = json.loads(response.text)
df = pd.DataFrame.from_dict(jsondata)
df.head()
```

Como podemos observar la informacion presentada de esta manera no nos esta diciendo nada, esto se debe a que lo que realmente de interes es la columna "value", por lo que se creara un nuevo dataframe con el contenido de esta columna. Pero no sin antes revisar si las otras columnas tienen algo informacion que nos pueda ser de utilidad.

```python
print(df["title"].nunique(),df["path"].nunique(),df["page"].nunique())
```
