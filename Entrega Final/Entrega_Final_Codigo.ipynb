# Modelos

Para comenzar con los modelos lo primero sera aplicar dos coeficientes de correlacion para poder poner en numeros lo que ya pudimos avistar con los graficos anteriores.

Los coeficientes que utilizaremos seran el de Pearson, ya que la mayoria de las variables parecen tener una relacion lineal con la liga, y el coeficiente de Spearman para las relaciones que no sean lineales, como lo parecia ser en el caso de "MinimapsAttacks"

```python
columns = data_skillcraft.columns
columns = columns.drop(["LeagueIndex"])
columns = list(columns)

for col in columns:
  coef_pearson = data_skillcraft[col].corr(data_skillcraft["LeagueIndex"])
  coef_corr_spearman = data_skillcraft[col].corr(data_skillcraft["LeagueIndex"], method='spearman')
  print("Los coeficientes de correlacion para la variable:" + col)
  print("Pearson: " + str(coef_pearson))
  print("Spearman: " + str(coef_corr_spearman))
```

Observando los resultados obtenidos podemos notar que la diferencia entre ambos coeficientes es minima, de todas formas solo tomaremos el valor de coeficiente mas alto dee cada variable para poder compararlas.

```python
correlaciones = pd.DataFrame(columns=['Variable', 'Coeficiente'])

for col in columns:
  coef_pearson = data_skillcraft[col].corr(data_skillcraft["LeagueIndex"])
  coef_corr_spearman = data_skillcraft[col].corr(data_skillcraft["LeagueIndex"], method='spearman')
  if abs(coef_pearson) < abs(coef_corr_spearman):
    correlaciones = correlaciones.append({'Variable': col, 'Coeficiente': coef_corr_spearman}, ignore_index=True)
  else:
    correlaciones = correlaciones.append({'Variable': col, 'Coeficiente': coef_pearson}, ignore_index=True)
```

Para finalizar lo ordenaremos por importancia y realizaremos un grafico mostrando las variables mas representativas

```python
correlaciones = correlaciones.reindex(correlaciones['Coeficiente'].abs().sort_values(ascending=True).index)

print(correlaciones)
```

```python
plt.barh(correlaciones['Variable'], correlaciones['Coeficiente'].abs())
plt.xlabel('Coeficiente de correlación')
plt.ylabel('Variable')
plt.title('Variables más importantes')

plt.show()
```

En conclusión, el análisis del gráfico revela que los factores más relevantes para el rendimiento de los jugadores se encuentran en habilidades intrínsecas, como el tiempo de reacción y las acciones por minuto, en lugar de aspectos biológicos como la edad. Contrariamente a lo esperado, el aumento de la edad no parece ser determinante a la hora de alcanzar un alto rendimineto en el juego pero es importante destacar que la muestra de datos utilizada en el análisis se compone principalmente de jugadores alrededor de los 20 años, con escasos registros de jugadores mayores de 30 años. Además, resulta sorprendente que elementos directamente relacionados con la jugabilidad, como la creación de unidades únicas o complejas, el uso de habilidades complejas o los clics en el mini mapa, sean considerados factores menos influyentes en el rendimiento de los jugadores. Estos hallazgos sugieren la importancia de desarrollar y entrenar habilidades específicas relacionadas con el tiempo de reacción y la toma de decisiones rápidas para alcanzar es vital para alcanzar un alto desempeño en los juegos analizados. # Optimizacion del modelo

A continuacion para mejorar el rendimiendto del modelo se aplicaran modelos de Boosting, Crosslvalidation y de seleccion de hiperparametros.

```python
#Volvemos a importar el dataset para asegurarnos que no se vio afectado por las operaciones anteriores

data_skillcraft = pd.read_csv(r'../SkillCraft.csv', delimiter = ",")
data_skillcraft = data_skillcraft.drop("GameID", axis=1)
Q1 = data_skillcraft.quantile(0.25)
Q3 = data_skillcraft.quantile(0.75)
IQR = Q3 - Q1

for columna in data_skillcraft.columns:
  data_skillcraft.loc[data_skillcraft[columna] > Q3[columna] + 1.5*IQR[columna] , columna] = Q3[columna] + 1.5*IQR[columna]

for i in range(1,6):
  data_skillcraft.loc[data_skillcraft.LeagueIndex==i, 'Apto'] = 0

  data_skillcraft.loc[data_skillcraft.LeagueIndex==6, 'Apto'] = 1
  data_skillcraft.loc[data_skillcraft.LeagueIndex==7, 'Apto'] = 1
```

```python
#Definimos las variables de entrada y salida y luego las normalizamos

X = data_skillcraft.drop(["LeagueIndex","Apto"], axis=1)
y = data_skillcraft['Apto']

scaler = StandardScaler()
x = scaler.fit_transform(X)
```

```python
#Definimos la cantidad de folds a realizar

num_folds = 5
skf = StratifiedKFold(n_splits=num_folds)
```

```python
#Instanciamos el modelo

modelo_base = RandomForestClassifier()
```

```python
#Determinamos entre que hipermaramotros iterar

param_distributions = {
    "n_estimators": [100, 250, 500, 1000],
    "max_depth": [None, 2, 5, 10],
    'min_samples_split':[5,10,50] ,
    'min_samples_leaf':[2,5,10]
}
```

```python
# Crear el objeto de búsqueda de hiperparámetros
halving_search = HalvingRandomSearchCV(
    modelo_base, param_distributions, factor=2, cv=skf, n_jobs=-1
)
```

```python
#Inicializamos las listas para almacenar los datos
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
variable_importances = []
```

```python
#Iteramos sobre cada fold, en cada caso se entrena el modelo y se busca optimizar los hiperparametros
for train_index, test_index in skf.split(X, y):
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    
    halving_search.fit(X_train, y_train)

    print(halving_search.best_params_)
    
    modelo_final = GradientBoostingClassifier(**halving_search.best_params_, learning_rate=0.1)

    
    modelo_final.fit(X_train, y_train)

    
    y_pred = modelo_final.predict(X_test)

    
    accuracy_scores.append(modelo_final.score(X_test, y_test))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))

   
    variable_importances.append(modelo_final.feature_importances_)
```

```python
#Calculamos las medias de las metricas obtenidas
mean_accuracy = np.mean(accuracy_scores)
mean_precision = np.mean(precision_scores)
mean_recall = np.mean(recall_scores)
mean_f1 = np.mean(f1_scores)
std_accuracy = np.std(accuracy_scores)
std_precision = np.std(precision_scores)
std_recall = np.std(recall_scores)
std_f1 = np.std(f1_scores)
mean_variable_importances = np.mean(variable_importances, axis=0)
nombres_variables = X.columns
```

```python
#Creamos un dataframe con la importancia de cada variable
df_importancias = pd.DataFrame({'Variable': nombres_variables, 'Coeficiente': mean_variable_importances})
df_importancias = df_importancias.sort_values('Coeficiente', ascending=True)
```

```python
print("Resultados de la validación cruzada Stratified K-Fold:")
print("Accuracy: {:.4f} +/- {:.4f}".format(mean_accuracy, std_accuracy))
print("Precision: {:.4f} +/- {:.4f}".format(mean_precision, std_precision))
print("Recall: {:.4f} +/- {:.4f}".format(mean_recall, std_recall))
print("F1-score: {:.4f} +/- {:.4f}".format(mean_f1, std_f1))
```

Podemos observar que, después de aplicar métodos de boosting, selección de hiperparámetros y folds, se obtuvo una ligera mejora en las métricas del modelo. Sin embargo, estas métricas indican un rendimiento deficiente del modelo en general. Esto puede atribuirse a varios factores, como la falta de datos suficientes o la posible complejidad de la predicción que se intenta alcanzar.

Es importante tener en cuenta que en juegos complejos como StarCraft 2, hay múltiples variables y aspectos que pueden influir en el rendimiento de los jugadores.
